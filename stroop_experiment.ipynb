# Die Stroop-Aufgabe

In einem Experiment zum Stroop-Effekt zeigen wir ein Wort (z.B. **ROT**) in einer bestimmten Schriftfarbe (z.B. blau) an.
Die Aufgabe: Nenne die **Schriftfarbe** und nicht das Wort!

Wenn Wort und Farbe nicht übereinstimmen (=inkongruenter Trial), macht das Gehirn mehr Fehler.
Das nennt man den **Stroop-Effekt**.

---

## Schritt 1: Daten vorbereiten
*(Bekannt: Variablen, Listen, Strings)*

Wir speichern die Trials als zwei parallele Listen.
Der Index bestimmt, welche Wort–Farbe-Paare zusammengehören.

```python
woerter = ["ROT",  "BLAU", "GRÜN", "ROT",  "GELB", "BLAU"]
farben  = ["rot",  "blau", "rot",  "blau", "gelb", "rot" ]

# Kongruent:   woerter[0]="ROT"  und farben[0]="rot"  → stimmt überein ✓
# Inkongruent: woerter[2]="GRÜN" und farben[2]="rot"  → stimmt NICHT überein ✗

print("Anzahl Trials:", len(woerter))
print("Erster Trial – Wort:", woerter[0], "| Farbe:", farben[0])
print("Trial kongruent (True) oder inkongruent (False)?", woerter[0] == farben[0])
```

Aber eigentlich sollte es doch ein kongruenter Trial sein, da "ROT" und "rot" gleich sind. Diese String-Methode lower() haben wir in Woche 2 kennengelernt.

```python
(woerter[0].lower()) == farben[0]
```

## Schritt 2: Einen einzigen Trial zeigen
*(Bekannt: Variablen, Listenindex, if/else)*

Wir starten mit Trial Nr. 0, ganz ohne Schleife.
So sehen wir das Grundprinzip, bevor wir es automatisieren.

```python
i = 0

wort  = woerter[i]
farbe = farben[i]
antwort = "rot"


print(f"Wort: {wort}  |  Schriftfarbe: {farbe}")

if antwort == farbe:
    print("Richtig!")
else:
    print(f"Falsch! Die Farbe war: {farbe}")
```

---

## Schritt 3: Alle Trials mit einer Schleife

Die Schleife macht genau das Gleiche wie Schritt 2,
nur automatisch für jeden Trial statt `i = 0` fest einzutippen.

```python
import random
antworten  = ["rot",  "blau", "rot",  "blau", "gelb", "rot" ]
random.shuffle(antworten)
print(antworten)
```

```python
len(woerter)
```

```python
richtig = 0
falsch  = 0

for i in range(len(woerter)):

    wort  = woerter[i]
    farbe = farben[i]
    antwort = antworten[i]
    
    print(f"\n--- Trial {i + 1} ---")
    print(str(antwort))

    if antwort == farbe:
        print("Richtig!")
        richtig = richtig + 1
    else:
        print(f"Falsch! Es war: {farbe}")
        falsch = falsch + 1
```

---

## Schritt 4: Auswertung
*(Bekannt: Variablen, Arithmetik, if/else)*

Jetzt schauen wir, ob der Stroop-Effekt messbar ist. Das speichern wir in den Variablen richtig und falsch, das könnten wir später beispielsweise in ein Logfile schreiben.

```python
print("════════ ERGEBNISSE ════════")
print(f"Richtig: {richtig}")
print(f"Falsch:  {falsch}")

if richtig > falsch:
    print("Gut gemacht!")
else:
    print("Der Stroop-Effekt hat dich erwischt!")
```
